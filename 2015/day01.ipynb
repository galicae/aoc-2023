# Day 01 - LiSP?

(: go up one, ): go down one, should be easy

```python
from util import file_to_list
```

```python
def navigate(s):
    floor = 0
    for c in s:
        if c == '(':
            floor +=1
        else:
            floor -= 1
    return floor
```

```python
assert navigate('(())') == 0
assert navigate('()()') == 0
assert navigate('(((') == 3
assert navigate('(()(()(') == 3
assert navigate('))(((((') == 3
assert navigate('())') == -1
assert navigate('))(') == -1
assert navigate(')))') == -3
assert navigate(')())())') == -3
```

```python
instructions = file_to_list('data/day01.txt')[0]
navigate(instructions)
```

```python
def to_the_basement(s):
    floor = 0
    for i, c in enumerate(s):
        if c == '(':
            floor +=1
        else:
            floor -= 1
        if floor == -1:
            break
    return i+1
```

```python
to_the_basement(instructions)
```

```python
def parse_range(s):
    a, b = s.split('-')
    a, b = int(a), int(b)
    return a, b

def read_day5(path):
    raw = file_to_list(path)
    for i, line in enumerate(raw):
        if line == '':
            break
    available = [int(s) for s in raw[i+1:]]
    fresh = [list(parse_range(s)) for s in raw[:i]]
    return fresh, available
```

```python
def in_range(item, fresh_range):
    start, stop = fresh_range
    return (item >= start) & (item <= stop)
```

```python
fresh, available = read_day5('./data/day05_test.txt')
fresh_count = 0
for ingr in available:
    intr = int(ingr)
    for r in fresh:
        if in_range(intr, r):
            # print(f'{intr} in range {r}')
            fresh_count += 1
            break

print(fresh_count)
```

```python
def overlaps(r1, r2):
    r1_smaller = (r1[0] <= r2[0]) & (r1[1] >= r2[0])
    r2_smaller = (r2[0] <= r1[0]) & (r2[1] >= r1[0])
    if r1_smaller or r2_smaller:
        return True
    return False
```

```python
assert overlaps([7, 10], [2, 5]) is False
assert overlaps([1, 10], [2, 5]) is True
assert overlaps([1, 10], [5, 10]) is True
assert overlaps([1, 10], [11, 20]) is False
assert overlaps([1, 10], [1, 10]) is True # this is what I was doing wrong...
```

```python
def merge(r1, r2):
    start = min(r1[0], r2[0])
    stop = max(r1[1], r2[1])
    return [start, stop]
```

```python
def total_fresh(ranges):
    count = 0
    for r in ranges:
        count += r[1] - r[0] + 1
    return count
```

```python
def merge_overlapping(ranges):
    non_overlapping = []
    while len(ranges) > 0:
        merged = False
        current_range = ranges.pop()
        # print(current_range)
        for i, r in enumerate(ranges):
            if overlaps(current_range, r):
                m = merge(current_range, r)
                # print(f'merging {current_range} with {r} into {m}')
                ranges.pop(i)
                ranges.append(m)
                merged = True
                break
        
        if not merged:
            non_overlapping.append(current_range)
        
        # print(fresh)

    return non_overlapping
```

```python
fresh, available = read_day5('./data/day05_test.txt')
non_overlapping = merge_overlapping(fresh)
total_fresh(non_overlapping)
```

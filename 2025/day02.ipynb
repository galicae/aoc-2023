```python
def invalid_id(s):
    middle = len(s) // 2
    return s[:middle] == s[middle:]
```

```python
invalid_id('1212')
```

```python
def examine_range(s, test_invalid):
    start, stop = s.split('-')
    start, stop = int(start), int(stop)
    list_invalid_ids = []
    for i in range(start, stop+1):
        # we could technically skip all numbers with an odd number of digits,
        # no clue if that really speeds it up?
        if test_invalid(str(i)):
            list_invalid_ids.append(i)
    return list_invalid_ids
```

```python
examine_range('1698522-1698528', invalid_id)
```

```python
invalid_ids = []
for r in id_ranges:
    invalid_ids += examine_range(r, invalid_id)

sum(invalid_ids)
```

## Part 2:

invalid IDs are now all IDs that consist of any repeating pattern, not just one repeat

```python
def hard_invalid_id(s):
    middle = len(s) // 2
    for i in range(1, middle+1):
        pattern = s[:i]
        repeats = len(s) // len(pattern)
        composite = pattern * repeats
        if composite == s:
            return True
    return False
```

```python
hard_invalid_id('123123123')
```

```python
examine_range('824824821-824824827', hard_invalid_id)
```

```python
def get_homework(path):
    raw = file_to_list(path)
    operations = raw.pop().split() # the last element
    N = len(raw[0].split())
    numbers = [[''] for n in range(N)]

    for line in raw:
        fragments = line.split()
        for i, f in enumerate(fragments):
            numbers[i].append(int(f))

    for h in numbers:
        h.pop(0)

    return numbers, operations
```

```python
def product(numbers):
    prod = 1
    for n in numbers:
        prod *= n
    return prod
```

```python
def do_homework(numbers, operations):
    grand_total = 0
    operation = {'+': sum, '*': product}
    for num, op in zip(numbers, operations):
        # print(num, op, operation[op])
        grand_total += operation[op](num)
    return grand_total
```

```python
numbers, operations = get_homework('./data/day06_test.txt')
do_homework(numbers, operations)
```

```python
numbers, operations = get_homework('./data/day06.txt')
do_homework(numbers, operations)
```

## part 2

cephalopods read top to bottom

```python
def get_homework_ceph(path):
    raw = file_to_list(path, raw=True)
    operations = raw.pop().split()
    N = len(raw[0])
    transposed = [''] * N
    for n in range(N):
        for line in raw:
            transposed[n] += line[n]

    numbers = []
    current = []
    for n in transposed:
        processed = n.replace(' ', '')
        if len(processed) != 0:
            current.append(int(processed))
        else:
            numbers.append(current)
            current = []
    numbers.append(current)

    return numbers, operations
```

```python
numbers, operations = get_homework_ceph('./data/day06_test.txt')
do_homework(numbers, operations)
```

```python
numbers, operations = get_homework_ceph('./data/day06.txt')
do_homework(numbers, operations)
```

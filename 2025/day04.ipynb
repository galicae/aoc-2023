```python
def adjacent(m, x, y):
    """given a map and a position, return the (up to) eight adjacent positions
    """
    M = len(m) # number of rows
    N = len(m[0]) # number of columns
    adjacent_positions = []
    for i in range(-1, 2):
        for j in range(-1, 2):
            # skip the middle square
            if i == 0 and j == 0:
                continue
            xi = x + i
            yj = y + j
            out_of_bounds = (xi >= M) | (yj >= N) | (xi < 0) | (yj < 0)
            if not out_of_bounds:
                adjacent_positions.append([xi, yj])
    return adjacent_positions
```

```python
adjacent(test, 1, 1)
```

```python
def is_reachable(m, x, y, occupied='@', empty='.', max_occupancy=4):
    if m[x][y] == empty:
        return False
    occupancy = 0
    for pos in adjacent(m, x, y):
        i, j = pos
        if m[i][j] == occupied:
            occupancy += 1

    if occupancy < max_occupancy:
        return True
    return False
```

```python
is_reachable(test, 0, 4)
```

```python
def find_reachable(m):
    reachable = []
    for i, _ in enumerate(m):
        for j, _ in enumerate(m[0]):
            if is_reachable(m, i, j):
                reachable.append([i, j])
    return reachable

def count_reachable(m):
    return len(find_reachable(m))
```

editing the map - this is where this becomes exciting!

```python
text = '#'.join(test)
```

```python
def map_to_str(i, j, M):
    return i*(M+1) + j

def remove_roll(s, i, j, M, empty='.'):
    pos = map_to_str(i, j, M)
    tmp = s[:pos] + empty + s[(pos+1):]
    return tmp
```

```python
def find_and_remove(m):
    total_rolls_removed = 0
    M = len(m) # number of rows
    reachable = find_reachable(m)
    text = '#'.join(m)
    while len(reachable) > 0:
        for pos in reachable:
            i, j = pos
            text = remove_roll(text, i, j, M)
        total_rolls_removed += len(reachable)
        reachable = find_reachable(text.split('#'))
    return total_rolls_removed
```

```python
from util import file_to_list
```

```python
rotations = file_to_list('./data/day01.txt')
```

```python
def rotate(dial, r, p=100):
    direction = r[0]
    steps = int(r[1:])
    if direction == 'L':
        dial -= steps
    else:
        dial += steps
    return dial % p
```

## part 1

```python
dial = 50
password = 0
for r in rotations:
    dial = rotate(dial, r)
    if dial == 0:
        password += 1
print(password)
```

## part 2

we now need to count the times we go _over_ the 0 in the dial; this presumably means we go from modulo to integer division.

```python
rotations = file_to_list('./data/day01_test.txt')
```

```python
def origin_passes(dial, r, p=100):
    direction = r[0]
    steps = int(r[1:])
    if direction == 'L':
        dial -= steps
    else:
        dial += steps
    origins = abs(dial // p)
    dial = dial % p
    return dial, origins
```

```python
dial = 50
password = 0
for r in rotations:
    dial, origins = origin_passes(dial, r)
    password += origins
print(password)
```

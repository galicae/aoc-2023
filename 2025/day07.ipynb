```python
def find_splits(incoming, line):
    # print(incoming)
    outgoing = []
    splits = 0
    tachyoned = line
    # check whether we can split
    for tachyon in incoming:
        if line[tachyon] == '.':
            if tachyon not in outgoing:
                outgoing.append(tachyon)
            tachyoned = tachyoned[:tachyon] + '|' + tachyoned[tachyon+1:]
        else:
            # split
            splits += 1
            # generate new tachyons
            left = tachyon - 1
            right = tachyon + 1
            # check margins:
            if left > -1 and left not in outgoing:
                tachyoned = tachyoned[:left] + '|' + tachyoned[left+1:]
                outgoing.append(left)
            if right < len(line) and right not in outgoing:
                tachyoned = tachyoned[:right] + '|' + tachyoned[right+1:]
                outgoing.append(right)
    
    # print(tachyoned)
    return splits, outgoing
```

```python
def find_total_splits(manifold):
    start = find_start(manifold)
    incoming = [start]
    total_splits = 0
    for line in manifold[1:]:
        splits, outgoing = find_splits(incoming, line)
        total_splits += splits
        incoming = outgoing

    return total_splits
```

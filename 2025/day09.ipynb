```python
def find_red_tiles(floor):
    red_tiles = []
    for row in floor:
        i, j = [int(x) for x in row.split(',')]
        red_tiles.append([i, j])
    return red_tiles
```

```python
def area(x1, y1, x2, y2):
    x = abs(x1-x2) + 1
    y = abs(y1-y2) + 1
    return x * y
```

```python
assert area(2, 5, 9, 7) == 24
assert area(7, 1, 11, 7) == 35
assert area(7, 3, 2, 3) == 6
```

largest rectangle: find all hashtags, try out all combinations

```python
def biggest_red_rectangle(floor):
    red_tiles = find_red_tiles(floor)
    biggest = 0
    for i, r1 in enumerate(red_tiles):
        for r2 in red_tiles[i+1:]:
            current = area(r1[0], r1[1], r2[0], r2[1])
            if current > biggest:
                biggest = current
    return biggest
```

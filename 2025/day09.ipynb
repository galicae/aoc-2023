```python
def find_red_tiles(floor):
    red_tiles = []
    for row in floor:
        i, j = [int(x) for x in row.split(',')]
        red_tiles.append([i, j])
    return red_tiles
```

```python
def area(x1, y1, x2, y2):
    x = abs(x1-x2) + 1
    y = abs(y1-y2) + 1
    return x * y
```

```python
assert area(2, 5, 9, 7) == 24
assert area(7, 1, 11, 7) == 35
assert area(7, 3, 2, 3) == 6
```

largest rectangle: find all hashtags, try out all combinations

```python
def biggest_red_rectangle(floor):
    red_tiles = find_red_tiles(floor)
    biggest = 0
    for i, r1 in enumerate(red_tiles):
        for r2 in red_tiles[i+1:]:
            current = area(r1[0], r1[1], r2[0], r2[1])
            if current > biggest:
                biggest = current
    return biggest
```

### part 2 

consecutive red tiles are connected by green tiles, all tiles enclosed by the perimeter thus formed
are also green. If we only allow red corners and green filling, which rectangles are now allowed?

if a rectangle is completely green, then that means that there must exist a red tile such that
the corners of the rectangle are less/equal to that tile.

```python
def upper_left(r1, r2, red_tiles):
    # find what the corners of this rectangle are
    x = min(r1[0], r2[0])
    y = min(r1[1], r2[1])
    for tile in red_tiles:
        if tile[0] <= x and tile[1] <= y:
            return True
    return False

def upper_right(r1, r2, red_tiles):
    # find what the corners of this rectangle are
    x = min(r1[0], r2[0])
    y = max(r1[1], r2[1])
    for tile in red_tiles:
        if tile[0] <= x and tile[1] >= y:
            return True
    return False

def lower_left(r1, r2, red_tiles):
    # find what the corners of this rectangle are
    x = max(r1[0], r2[0])
    y = min(r1[1], r2[1])
    for tile in red_tiles:
        if tile[0] >= x and tile[1] <= y:
            return True
    return False

def lower_right(r1, r2, red_tiles):
    # find what the corners of this rectangle are
    x = max(r1[0], r2[0])
    y = max(r1[1], r2[1])
    for tile in red_tiles:
        if tile[0] >= x and tile[1] >= y:
            return True
    return False

def is_not_green(r1, r2, red_tiles):
    if upper_left(r1, r2, red_tiles):
        if upper_right(r1, r2, red_tiles):
            if lower_left(r1, r2, red_tiles):
                if lower_right(r1, r2, red_tiles):
                    return False
    return True
```

```python
def biggest_green_rectangle(floor):
    red_tiles = find_red_tiles(floor)
    biggest = 0
    for i, r1 in enumerate(red_tiles):
        for r2 in red_tiles[i+1:]:
            current = area(r1[0], r1[1], r2[0], r2[1])
            if is_not_green(r1, r2, red_tiles):
                break
            if current > biggest:
                biggest = current
    return biggest
```

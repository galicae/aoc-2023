We are essentially building two-digit integers; the greatest possible integer we can build while
respecting reading order will have to start from the greatest available digit in the first len-1
positions.

Ties don't matter in the first run, for sure.

```python
def find_first_maximal_digit(s):
    max_value = 0
    max_index = -1
    for i, v in enumerate(s):
        num = int(v)
        if num > max_value:
            max_index = i
            max_value = num
    return str(max_value), max_index
```

```python
def maximum_joltage(bank):
    battery_1, battery_1_pos = find_first_maximal_digit(bank[:-1])
    battery_2, _ = find_first_maximal_digit(bank[battery_1_pos+1:])
    return int(battery_1 + battery_2)
```

```python
maximum_joltage('818181911112111')
```

```python
total_joltage = 0
for b in banks:
    total_joltage += maximum_joltage(b)

total_joltage
```

## Part 2

looks like we can repurpose part 1, we just have to be careful about how many digits we are allowed
to look into? Since this is base-10, it doesn't really matter if we have ties, we can always pick
the earliest battery with higher joltage. We might have to be careful with the indices though?

```python
def turn_it_to_twelve(bank):
    bt_prev = 0
    joltage = ''
    for i in range(12):
        offset = -(11 - i)
        if offset == 0: offset = len(bank) + 1
        # print(i, f'bank[{bt_prev}:{-offset}]', bank[bt_prev:offset])
        battery_i, bt_pos = find_first_maximal_digit(bank[bt_prev:offset])
        bt_prev += bt_pos + 1
        joltage += battery_i
        # print(battery_i, '#####')
    return int(joltage)
```

```python
turn_it_to_twelve('811111111111119')
```

```python
banks = file_to_list('data/day03_test.txt')
total_joltage = 0
for b in banks:
    joltage = turn_it_to_twelve(b)
    total_joltage += joltage
    print(b, joltage)

print(total_joltage)
```
